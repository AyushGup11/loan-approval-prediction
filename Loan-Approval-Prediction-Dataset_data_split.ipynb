```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("D:\Data Science\Projects\Loan Prediction Dataset\loan_approval_dataset.csv")

data.columns
```

```python
x = data.drop(" loan_status", axis=1)  # Replace with the actual column name you found
y = data[" loan_status"]
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42, stratify=y)

print("Training data shape:", x_train.shape)
print("Training data shape:", x_test.shape)
```

```python
train_data = pd.concat([x_train, y_train], axis=1)
test_data = pd.concat([x_test, y_test], axis=1)

train_data.to_csv("train.csv", index=False)
test_data.to_csv("test.csv", index=False)
```
